```python
import pandas as pd
import glob

# Use glob to find all Excel files that start with "stocks_report"
file_pattern = "stocks_report*.xlsx"
file_paths = glob.glob(file_pattern)

# Dictionary to store DataFrames for each file
stock_dataframes = {}

for file_path in file_paths:
    # Read the Excel file without setting the header
    df = pd.read_excel(file_path, sheet_name='Товары', header=None)

    # Drop the first, second, and fourth rows (0-based index: 0, 1, and 3)
    df = df.drop([0, 1, 3])

    # Set the new header to be the first row of the remaining DataFrame
    df.columns = df.iloc[0]

    # Drop the row that was used as the header
    df = df.drop(2).reset_index(drop=True)

    # Store the DataFrame in the dictionary with a descriptive key
    file_name = file_path.split('/')[-1]  # Extract the file name from the path
    stock_dataframes[file_name] = df

# Access each DataFrame individually
for file_name, df in stock_dataframes.items():
    print(f"DataFrame for {file_name}:")
    print(df.head())
    print("\n")

```
